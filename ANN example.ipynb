In a bioligical ANN, the following situation occurs:

Learning occurs by repeatedly activating certain neural connections over others, and this reinforces those connections.  This makes them more likely to produce a desired outcome given a specified input.  This learning involves feedback – when the desired outcome occurs, the neural connections causing that outcome become strengthened.

## The structure of an ANN

The biological neuron is simulated in an ANN by an activation function. In classification tasks (e.g. identifying spam e-mails) this activation function has to have a “switch on” characteristic – in other words, once the input is greater than a certain value, the output should change state i.e. from 0 to 1, from -1 to 1 or from 0 to >0. This simulates the “turning on” of a biological neuron. A common activation function that is used is the sigmoid function

```python
import matplotlib.pylab as plt
import numpy as np
import matplotlib 
plt.style.use('ggplot')

%matplotlib inline
x = np.arange(-10, 10, 0.01)
f = 1 / (1 + np.exp(-x))
plt.plot(x, f)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.show()
```

The function can be said to be activated with its value becomes equal to one

Nodes in ANN (also known as perceptrons) take a vector of real inputs, add them in a linear combination using a series of weights - which control the relative importance of the inputs - and then output a 1 or a -1 depending on whether the combined value passes some threshold, typically determined by a function that looks simiar to that above

The effects of adjusting weights in a simple model are shown below

```python
w1 = 0.5
w2 = 1.0
w3 = 2.0
l1 = 'w = 0.5'
l2 = 'w = 1.0'
l3 = 'w = 2.0'
for w, l in [(w1, l1), (w2, l2), (w3, l3)]:
    f = 1 / (1 + np.exp(-x*w))
    plt.plot(x, f, label=l)
plt.xlabel('x')
plt.ylabel('h_w(x)')
plt.legend(loc=2)
plt.show()
```

We can easily imagine f becoming more complicated, where we are combining multiple weights in this threshold function

Adding a bias allows us to make the function into an IF statement, because it shifts the location of the increase to 1

```python
w = 5.0
b1 = -8.0
b2 = 0.0
b3 = 8.0
l1 = 'b = -8.0'
l2 = 'b = 0.0'
l3 = 'b = 8.0'
for b, l in [(b1, l1), (b2, l2), (b3, l3)]:
    f = 1 / (1 + np.exp(-(x*w+b)))
    plt.plot(x, f, label=l)
plt.xlabel('x')
plt.ylabel('h_wb(x)')
plt.legend(loc=2)
plt.show()
```
